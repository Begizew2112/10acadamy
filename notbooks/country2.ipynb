```python
import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np

```

```python
#upload the csv file 
df = pd.read_csv('sierraleone-bumbuna.csv')
df.head()
```

```python
#the total shape of the uncleaned data
df.shape 
```

```python
#check if there is null value if the given columns
df.isnull().sum() 
```

```python
#to get the necessary information the uncleaned data
df.describe()
```

```python
#then clean the negative value of the GHI ,DNI,DHI
df1 = df[df['GHI']>=0]
df1.shape 
```

```python
# clean if there is still negative DNI,DHI  in cleaned GHI
```

```python
df22 = df1[df1['DNI']>=0]
df22.shape 
```

```python
df2 = df22[df22['DHI']>=0]
df2.shape 
```

```python
#next we check if there is any outlier on ModeA and ModeB using IQR
# Calculate Q1 (25th percentile) and Q3 (75th percentile)
Q1 = df2['ModA'].quantile(0.25)
Q3 = df2['ModA'].quantile(0.75)
IQR = Q3 - Q1

# Identify outliers
outliers = df2[(df2['ModA'] < Q1 - 1.5 * IQR) | (df2['ModA'] > Q3 + 1.5 * IQR)]
outliers # the result show empty data frame means that there is no outliers
```

```python
# Determine the bounds for outliers to drop 
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Drop outliers
df3 = df2[(df['ModA'] >= lower_bound) & (df2['ModA'] <= upper_bound)]

# Verify the result
df3.shape # look the shape since there are two outliers it decrease the shape by two 
```

```python
# Calculate Q1 (25th percentile) and Q3 (75th percentile)
Q11 = df3['ModB'].quantile(0.25)
Q33 = df3['ModB'].quantile(0.75)
IQR = Q33 - Q11

# Identify outliers
outliers = df3[(df3['ModB'] < Q11 - 1.5 * IQR) | (df3['ModB'] > Q33 + 1.5 * IQR)]
outliers # the result show only one outliers 
```

```python
# Determine the bounds for outliers to drop 
lower_bound = Q11 - 1.5 * IQR
upper_bound = Q33 + 1.5 * IQR

# Drop outliers
df4 = df3[(df['ModB'] >= lower_bound) & (df3['ModB'] <= upper_bound)]

# Verify the result
df4.shape # look the shape since theere is one outliers it decrease the shape by one
```

```python
# to check and clean up the outliers in the ws 
Q1 = df4['WS'].quantile(0.25)
Q3 = df4['WS'].quantile(0.75)
IQR = Q3 - Q1

# Identify outliers
outliers = df4[(df4['WS'] < Q1 - 1.5 * IQR) | (df4['WS'] > Q3 + 1.5 * IQR)]
outliers.shape # the result show there  is an outliers so we drop it
```

```python
# Determine the bounds for outliers to drop 
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Drop outliers
df5 = df4[(df4['WS'] >= lower_bound) & (df4['WS'] <= upper_bound)]

# Verify the result
df5.shape
```

```python
238869 - 685 # our cleaning is correct 
```

```python
Q1 = df5['WSgust'].quantile(0.25)
Q3 = df5['WSgust'].quantile(0.75)
IQR = Q3 - Q1

# Identify outliers
outliers = df5[(df5['WSgust'] < Q1 - 1.5 * IQR) | (df5['WSgust'] > Q3 + 1.5 * IQR)]
outliers.shape 
```

```python
# Determine the bounds for outliers and drop the outlier
l_bound = Q1 - 1.5 * IQR
up_bound = Q3 + 1.5 * IQR

# Drop outliers
df6= df5[(df5['WSgust'] >= l_bound) & (df5['WSgust'] <= up_bound)]

# Verify the result
df6.shape
```
