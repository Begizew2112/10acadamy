```python
import pandas as pd

# Load summary statistics from CSV files
summary_region1 = pd.read_csv('summary_region1.csv')
summary_region2 = pd.read_csv('summary_region2.csv')
summary_region3 = pd.read_csv('summary_region3.csv')
```

```python
# Display the summaries for verification
print("Summary Statistics for Region 1:")
summary_region1
```

```python
print("\nSummary Statistics for Region 2:")
summary_region2
```

```python
print("\nSummary Statistics for Region 3:")
summary_region3
```

```python
import matplotlib.pyplot as plt


# Combine the summary statistics into one DataFrame for easier plotting
combined_summary = pd.concat([summary_region1.set_index('Unnamed: 0'), 
                              summary_region2.set_index('Unnamed: 0'), 
                              summary_region3.set_index('Unnamed: 0')], 
                             axis=1)

# Rename columns for clarity
combined_summary.columns = ['Region 1_GHI', 'Region 1_DNI', 'Region 1_DHI', 'Region 1_Tamb',
                            'Region 2_GHI', 'Region 2_DNI', 'Region 2_DHI', 'Region 2_Tamb',
                            'Region 3_GHI', 'Region 3_DNI', 'Region 3_DHI', 'Region 3_Tamb']

# Plot bar charts to compare mean values for each metric across the regions

# GHI comparison
combined_summary.loc['mean'][['Region 1_GHI', 'Region 2_GHI', 'Region 3_GHI']].plot(kind='bar', color=['blue', 'green', 'red'])
plt.title('Mean GHI Comparison Across Regions')
plt.ylabel('Mean GHI')
plt.xticks(rotation=0)
plt.show()

# DNI comparison
combined_summary.loc['mean'][['Region 1_DNI', 'Region 2_DNI', 'Region 3_DNI']].plot(kind='bar', color=['blue', 'green', 'red'])
plt.title('Mean DNI Comparison Across Regions')
plt.ylabel('Mean DNI')
plt.xticks(rotation=0)
plt.show()

# DHI comparison
combined_summary.loc['mean'][['Region 1_DHI', 'Region 2_DHI', 'Region 3_DHI']].plot(kind='bar', color=['blue', 'green', 'red'])
plt.title('Mean DHI Comparison Across Regions')
plt.ylabel('Mean DHI')
plt.xticks(rotation=0)
plt.show()

# Tamb comparison
combined_summary.loc['mean'][['Region 1_Tamb', 'Region 2_Tamb', 'Region 3_Tamb']].plot(kind='bar', color=['blue', 'green', 'red'])
plt.title('Mean Temperature (Tamb) Comparison Across Regions')
plt.ylabel('Mean Temperature (Tamb)')
plt.xticks(rotation=0)
plt.show()
```
